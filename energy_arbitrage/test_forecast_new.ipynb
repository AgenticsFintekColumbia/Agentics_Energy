```python
import os, sys, warnings
from mcp import StdioServerParameters
from crewai_tools import MCPServerAdapter

warnings.filterwarnings("ignore")
os.environ["CREWAI_TOOLS_DISABLE_AUTO_INSTALL"] = "1"

params = StdioServerParameters(
    command=sys.executable,
    args=["-m", "agentic_energy.forecast.forecast_mcp_server"],
    env=os.environ,
)

async def test():
    with MCPServerAdapter(params) as tools:
        tool = [t for t in tools if t.name == "forecast_for_date"][0]
        call_fn = getattr(tool, "call", None) or getattr(tool, "run", None)
        
        result = call_fn(args={
            "date": "2019-06-15",
            "target": "prices",
            "model_type": "LSTM"
        })
        
        print(f"MAE: {result['mae']:.2f}")
        print(f"Predictions: {result['predictions'][:6]}")
        return result

result = await test()
```
